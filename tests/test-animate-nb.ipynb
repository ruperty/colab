```python
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from matplotlib import style
from random import random
from IPython.display import HTML
```

```python
x = random()

style.use('fivethirtyeight')

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlim(( 0, 1))
ax1.set_ylim((0, 1))


line, = ax1.plot([], [], 'ro')
```

```python
xs = []
ys = []
def init():
    line.set_data([], [])
    return (line,)
```

```python
def animate(i):
    #print(i)
    x = random()
    y = random()
    xs.append(float(x))
    ys.append(float(y))
    #ax1.clear()
    #ax1.plot(xs, ys, 'bo')
    line.set_data(xs, ys)
    return (line,)
```

```python
anim = animation.FuncAnimation(fig, animate, interval=100, frames=100, blit=True)
#print('ani')
#plt.show()
#print('show')
```
